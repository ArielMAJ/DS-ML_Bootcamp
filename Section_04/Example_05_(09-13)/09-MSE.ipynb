```python
# Notebook imports and packages
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt

%matplotlib inline
```

#  Mean Squared Error: a cost function for regression problems


$$RSS=\sum^n_{i=1}\left( y^{(i)}-h_{\theta}x^{(i)}\right)^2$$

---

$$MSE=\frac{1}{n}\sum^n_{i=1}\left( y^{(i)}-h_{\theta}x^{(i)}\right)^2$$

---

$$MSE=\frac{1}{n}\sum^n_{i=1}\left( y-\hat{y}\right)^2$$

```python
x = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose() # Shape: (7,1)
y = np.array([[1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]]).T # Shape: (7,1)
# "T" in "y.T" means "transpose". It does the same as "x.transpose()". 
# The same can be achieved with "x.reshape()".

x.shape, y.shape
```

```python
# "x" and "y" were written as they did because "regr.fit()" only accept 2D arrays.
regr = LinearRegression()
regr.fit(x, y)
print(f"Theta 0: {regr.intercept_[0]}\nTheta 1: {regr.coef_[0, 0]}") # \theta_0 and \theta_1

plt.scatter(x, y, s=50) # The actual data.
plt.plot(x, regr.predict(x), color='orange') # The linear regression.
plt.xlabel("x")
plt.ylabel('y')
plt.show()
```

---

$\hat{y} = \theta_0 + \theta_1x$:

```python
y_hat = regr.intercept_[0] + regr.coef_[0, 0] * x 
```

---

MSE function:

```python
def mse(y, y_hat):
    return np.average(np.power(y-y_hat,2), axis=0)
```

---

You can calculate MSE in many different ways. You can code up your own functions, but there are many packages with them already implemented for you as well.

```python
print(
    f"My MSE function:\t{mse(y,y_hat)[0]}",
    f"MSE package+my calcs:\t{mean_squared_error(y,y_hat)}",
    f"MSE pack w/ pred:\t{mean_squared_error(y, regr.predict(x))}",
    sep='\n'
)
```
