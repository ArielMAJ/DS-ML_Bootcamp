```python
# Notebook imports and packages
import numpy as np

np.set_printoptions(linewidth=400)
```

#  Finishing up lesson 50


```python
x = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).T
y = np.array([[1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]]).T
```

```python
def mse(y, y_hat):
    return np.average(np.power(y-y_hat,2), axis=0)
```

## My result for minimum cost

```python
nr_thetas = 2001
th_0 = th_1 = np.linspace(start=-2, stop=4, num=nr_thetas)
plot_t0, plot_t1 = np.meshgrid(th_0, th_1)

print(plot_t0,'\n\n')
print(plot_t1, '\n\n')
```

```python
plot_cost = np.zeros((nr_thetas, nr_thetas))

for i in range(nr_thetas):
    for j in range(nr_thetas):
        y_hat = plot_t0[i][j] + plot_t1[i][j]*x
        plot_cost[i][j] = mse(y, y_hat)

print(plot_cost)
```

```python
ij_min = np.unravel_index(indices=plot_cost.argmin(), shape=plot_cost.shape)
print(
    f"Minimum value of plot_cost: {plot_cost.min()}",
    f"Position of minimum: {ij_min}",
    f"Best Theta 0: {plot_t0[ij_min]}",
    f"Best Theta 1: {plot_t1[ij_min]}",
    sep='\n'
)
```
