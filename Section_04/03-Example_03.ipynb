# Example 03 - Divergence and Overflow

$$h(x)=x^5-2x^4+2$$
$$\mathfrak{d}h(x)=5x^4-8x^3$$

```python
x_3 = np.linspace(start=-2.5, stop=2.5, num=2000) # Generates "data"

def h(x):
    return x**5 - 2*x**4 + 2

def dh(x):
    return 5*x**4 - 8*x**3
```

```python
local_min, x_list, deriv_list = gradient_descent(dh, .2)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12,4))

# Cost function chart:
axs[0].plot(x_3, h(x_3), color="blue", linewidth=3, alpha=.8)
axs[0].set_title("Cost function:")
axs[0].set_xlim([-2.5, 2.5])
axs[0].set_ylim(-1, 4)
axs[0].grid()
axs[0].set_ylabel("h(x)", fontsize=16)
axs[0].scatter(x_list, h(x_list), color='red', s=50, alpha=.3)

# Derivative chart:
axs[1].set_title("Derivative:")
axs[1].plot(x_3, dh(x_3), color='skyblue', linewidth=5, alpha=.8)
axs[1].set_xlabel("x", fontsize=16)
axs[1].set_ylabel("dh(x)", fontsize=16)
axs[1].set_xlim([-2.5, 2.5])
axs[1].set_ylim(-4, 5)
axs[1].grid()
axs[1].scatter(x_list, deriv_list, color='red', s=50, alpha=.3)


plt.show()
```

```python
local_min, x_list, deriv_list = gradient_descent(dh, -.2, multiplier=.02, max_iter=72)

fig, axs = plt.subplots(1, 2, figsize=(12,4))

# Cost function chart:
axs[0].plot(x_3, h(x_3), color="blue", linewidth=3, alpha=.8)
axs[0].set_title("Cost function:")
axs[0].set_xlim([-2.5, 2.5])
axs[0].set_ylim(-1, 4)
axs[0].grid()
axs[0].set_ylabel("h(x)", fontsize=16)
axs[0].scatter(x_list, h(x_list), color='red', s=50, alpha=.3)

# Derivative chart:
axs[1].set_title("Derivative:")
axs[1].plot(x_3, dh(x_3), color='skyblue', linewidth=5, alpha=.8)
axs[1].set_xlabel("x", fontsize=16)
axs[1].set_ylabel("dh(x)", fontsize=16)
axs[1].set_xlim([-2.5, 2.5])
axs[1].set_ylim(-4, 5)
axs[1].grid()
axs[1].scatter(x_list, deriv_list, color='red', s=50, alpha=.3)

print(local_min, h(local_min))
plt.show()
```
