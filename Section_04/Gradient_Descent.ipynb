# Notebook imports and Packages

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example 01

$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
x_1 = np.linspace(start=-3, stop=3, num=61, endpoint=True)
```

```python
x_1
```

```python
plt.xlim([-3, 3])
plt.ylim(0,8)
plt.xlabel("x", fontsize=16)
plt.ylabel("f(x)", fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Slope & derivatives

```python
def df(x):
    return 2 * x + 1
```

```python
from matplotlib import lines
plt.figure(figsize=[6, 8])

# Cost function chart:
plt.subplot(2, 1, 1)

plt.title("Cost function:")
plt.xlim([-3, 3])
plt.ylim(0,13)
plt.grid()
plt.ylabel("f(x)", fontsize=16)
plt.plot(x_1, f(x_1), color="blue", linewidth=3)

# Derivative chart:
plt.subplot(2, 1, 2)
plt.title("Derivative:")
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)
plt.ylabel("df(x)", fontsize=16)
plt.xlabel("x", fontsize=16)
plt.xlim([-3, 3])
plt.grid()

plt.show()
```

# Python Loops & Gradient Descent

```python
counter = 1
while counter < 10:
    for i in range(100):
        counter *= 1 + i/10000
    print(round(counter,3))
```

```python
x = 3
step_multiplier = .07

x_list = [x]
```

```python
for n in range(50):
    x = x - step_multiplier * df(x)
    x_list.append(x)

print(f"Local minimum:\t{x}")
print(f"Slope at x:\t{df(x)}")
print(f"f(x) cost at x:\t{f(x)}")
```

```python
from matplotlib import lines
# plt.figure(figsize=[10, 8])

fig, axs = plt.subplots(2, 2)

# Cost function chart:

axs[0, 0].plot(x_1, f(x_1), color="blue", linewidth=3, alpha=.8)
axs[0, 0].set_title("Cost function:")
axs[0, 0].set_xlim([-3, 3])
axs[0, 0].set_ylim(0,13)
axs[0, 0].grid()
axs[0, 0].set_ylabel("f(x)", fontsize=16)

axs[0, 0].scatter(x_list, f(np.array(x_list)), color='black', s=100, alpha=.1)

# # Derivative chart:
# plt.title("Derivative:")
# plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=.8)
# plt.ylabel("df(x)", fontsize=16)
# plt.xlabel("x", fontsize=16)
# plt.xlim([-3, 3])
# plt.grid()
# plt.scatter(x_list, df(np.array(x_list)), color='black', s=100, alpha=.1)

# from matplotlib import lines
# plt.figure(figsize=[6, 8])

# # Cost function chart:
# plt.subplot(1, 2)

# plt.title("Cost function:")
# plt.xlim([-.55, -.45])
# plt.grid()
# plt.ylabel("f(x)", fontsize=16)
# plt.plot(x_1, f(x_1), color="blue", linewidth=3, alpha=.8)

# plt.scatter(x_list, f(np.array(x_list)), color='black', s=100, alpha=.1)

# # Derivative chart:
# plt.subplot(2, 2)
# plt.title("Derivative:")
# plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=.8)
# plt.ylabel("df(x)", fontsize=16)
# plt.xlabel("x", fontsize=16)
# plt.xlim([-.55, -.45])
# plt.grid()
# plt.scatter(x_list, df(np.array(x_list)), color='black', s=100, alpha=.1)

# plt.draw

plt.show()
```
