# Example 02 - Multiple Minima vs Initial Guess & Advanced Functions

$$g(x)=x^4-4x^2+5$$
$$\mathfrak{d}g(x)=4x^3-8x$$

```python
x_2 = np.linspace(-2, 2, 1001)
x_2
```

```python
def g(x):
    return x**4-4*x**2+5

def dg(x):
    return 4*x**3-8*x
```

```python
# from matplotlib import lines
# plt.figure(figsize=[10, 8])

fig, axs = plt.subplots(1, 2, figsize=(12,4))

# Cost function chart:

axs[0].plot(x_2, g(x_2), color="blue", linewidth=3, alpha=.8)
axs[0].set_title("Cost function:")
axs[0].set_xlim([-2, 2])
# axs[0, 0].set_ylim(0,13)
axs[0].grid()
axs[0].set_ylabel("g(x)", fontsize=16)
# axs[0, 0].scatter(x_list, f(np.array(x_list)), color='black', s=50, alpha=.1)

# Derivative chart:
axs[1].set_title("Derivative:")
axs[1].plot(x_2, dg(x_2), color='skyblue', linewidth=5, alpha=.8)
axs[1].set_xlabel("x", fontsize=16)
axs[1].set_ylabel("dg(x)", fontsize=16)
axs[1].set_xlim([-2, 2])
axs[1].grid()
# axs[0, 1].scatter(x_list, dg(np.array(x_list)), color='black', s=50, alpha=.1)


plt.show()
```

## GD function

```python
def gradient_descent(derivative_fun, initial_guess, multiplier=0.01, precision=0.0001):
    x = initial_guess

    x_list = [x]

    for n in range(1, 500):
        x = x - multiplier * derivative_fun(x)
        x_list.append(x)
        if abs(x-x_list[-2]) < precision:
            break

    print(f"Local minimum:\t{x}")
    print(f"Number Steps:\t{len(x_list)}")
    print(f"Slope at x:\t{derivative_fun(x)}")
    # print(f"g(x) cost at x:\t{g(x)}")

    x_list = np.array(x_list)
    return x, x_list, derivative_fun(x_list)
```
