```python
# from IPython.display import clear_output as clear
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
ame_data = fetch_openml(name="house_prices", as_frame=True)
data = pd.DataFrame(data=ame_data.data, columns=ame_data.feature_names)
data["PRICE"] = ame_data.target/1000
```

# Visualising Data - Histograms, Distributions and Bar Charts

Color pallete: https://colorhunt.co/palette/e7e0c9c1cfc06b7aa111324d

```python
# plt.rcParams['text.color'] = plt.rcParams['axes.labelcolor'] = plt.rcParams['xtick.color'] = plt.rcParams['ytick.color'] = "#11324D"
plt.rcParams.update({
    'text.color' : "#11324D",
    'axes.labelcolor' : "#11324D",
    'xtick.color' : "#11324D",
    'ytick.color' : "#11324D"
})
```

```python
fig = plt.figure(figsize=(5, 3), facecolor='#E7E0C9', dpi=200)
# fig.set_facecolor(color='#E7E0C9')

ax = plt.axes()
for spine in ax.spines:
    ax.spines[spine].set_color("#11324D")
ax.set_facecolor(color='#C1CFC0')


plt.hist(data["PRICE"], bins=50, ec="#11324D", color="#6B7AA1")
plt.xlabel("Price (in thousands)")
plt.ylabel("Frequency")

plt.show()
```

```python
fig = plt.figure(figsize=(8, 3), dpi=200, facecolor='#E7E0C9')

ax = fig.add_subplot(1, 2, 1)
for spine in ax.spines:
    ax.spines[spine].set_color("#11324D")
ax.set_facecolor(color='#C1CFC0')

sns.distplot(data['PRICE'], bins=50, color="#6B7AA1", kde=False) #hist=False, kde=False


ax = fig.add_subplot(1, 2, 2)
for spine in ax.spines:
    ax.spines[spine].set_color("#11324D")
ax.set_facecolor(color='#C1CFC0')

sns.histplot(data['PRICE'], bins=50, color="#6B7AA1", kde=False) #hist=False, kde=False


plt.show()
```

```python
# Color pallete: https://colorhunt.co/palette/e7e0c9c1cfc06b7aa111324d

def paint_my(ax):
    for spine in ax.spines:
        ax.spines[spine].set_color("#11324D")
    ax.set_facecolor(color='#C1CFC0')

fig = plt.figure(figsize=(10, 3), facecolor='#E7E0C9', dpi=200)
# fig.set_facecolor(color='#E7E0C9')

ax = fig.add_subplot(1, 2, 1)
paint_my(ax)

plt.hist(data["TotRmsAbvGrd"], ec="#11324D", color="#6B7AA1")
plt.xlabel("TotRmsAbvGrd")
plt.ylabel("Frequency")


ax = fig.add_subplot(1, 2, 2)
paint_my(ax)

plt.hist(data["OverallQual"], ec="#11324D", color="#6B7AA1")
plt.xlabel("OverallQual")
plt.ylabel("Frequency")


plt.show()
```
